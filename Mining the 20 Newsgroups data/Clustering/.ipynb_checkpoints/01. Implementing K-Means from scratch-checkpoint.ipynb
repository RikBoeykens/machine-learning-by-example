Loading the iris dataset

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, 2:4]
y = iris.target
```

Plotting the data

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
y_0 = np.where(y==0)
plt.scatter(X[y_0, 0], X[y_0, 1])
y_1 = np.where(y==1)
plt.scatter(X[y_1, 0], X[y_1, 1])
y_2 = np.where(y==2)
plt.scatter(X[y_2, 0], X[y_2, 1])
plt.show()
```

STEP 1: Specifying K

```python
k = 3
```

#STEP 2: Initializing centroids

```python
random_index = np.random.choice(range(len(X)), k)
print(random_index)
centroids = X[random_index]
print(centroids)
```

Function to visualise centroids

```python
def visualize_centroids(X, centroids):
    plt.scatter(X[:, 0], X[:, 1])
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='#050505')
    plt.show()

visualize_centroids(X, centroids)
```

STEP 3: Assigning clusters to the nearest centroids

Function to get distances between two points

```python
def dist(a, b):
    return np.linalg.norm(a - b, axis = 1)
print(dist(X[1], centroids))
```

Function to assign a sample to the cluster of the nearest centroid

```python
def assign_cluster(X, centroids):
    distances = dist(X, centroids)
    cluster = np.argmin(distances)
    return cluster
print(assign_cluster(X[1], centroids))
```

STEP 4: Update the centroids to the mean of all samples in the individual cluster

```python
def update_centroids(X, centroids, clusters):
    for i in range(k):
        cluster_i = np.where(clusters == i)
        centroids[i] = np.mean(X[cluster_i], axis=0)
```

STEP 5: Repeat step 3 and 4 until model converges: centroids move small enough or sufficient iterations have been taken

```python
tol = 0.0001        # tolerance - how far the centroids have moved
max_iter = 100      # max iterations - maximum amount of iterations

iter = 0                      # initial iteration
centroids_diff = 100000       # initial centroid difference
clusters = np.zeros(len(X))   # holds cluster info for each value of X

from copy import deepcopy
while iter < max_iter and centroids_diff > tol:
    for i in range(len(X)):
        clusters[i] = assign_cluster(X[i], centroids)
    centroids_prev = deepcopy(centroids)
    update_centroids(X, centroids, clusters)
    
    iter += 1
    centroids_diff = np.linalg.norm(centroids - centroids_prev)
    print('Iteration:', str(iter))
    print('Centroids:\n', centroids)
    print('Centroids move: {:5.4f}'.format(centroids_diff))
    visualize_centroids(X, centroids)
```

Plotting the clusters

```python
for i in range(k):
    cluster_i = np.where(clusters == i)
    plt.scatter(X[cluster_i, 0], X[cluster_i, 1])
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='#050505')
plt.show()
```
